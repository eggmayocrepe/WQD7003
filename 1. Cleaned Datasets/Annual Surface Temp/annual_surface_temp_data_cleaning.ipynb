```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.float_format', '{:.3f}'.format)
```

**Dataset Overview**

```python
# Read the CSV file
df = pd.read_csv("Annual_Surface_Temperature_Change.csv")
df
```

```python
# Check the column names
print(df.columns)
```

```python
# Select the relevant columns and create a new DataFrame
Sel_cols = ['Country', 'ISO2', 'ISO3', 'F2000', 'F2001', 'F2002',
       'F2003', 'F2004', 'F2005', 'F2006', 'F2007', 'F2008', 'F2009', 'F2010',
       'F2011', 'F2012', 'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018',
       'F2019', 'F2020', 'F2021', 'F2022']
new_df = df[Sel_cols]
new_df
```

```python
# Summary statistics of the new DataFrame
new_df.describe()
```

```python
# Check the NA for each column
new_df.isna().sum()
```

```python
# Separate the columns into two groups for plotting
columns1 = new_df.columns[3:14]
columns2 = new_df.columns[14:26]
```

```python
# First group of columns (2000 to 2010)
plt.Figure(figsize=(20, 10))
new_df[columns1].plot(kind='box', figsize=(15, 6), title='Annual Surface Temperature Change from 2000 to 2010')
plt.tight_layout()
plt.show()
```

```python
# Second group of columns (2011 to 2022)
new_df[columns2].plot(kind='box', title='Annual Surface Temperature Change from 2011 to 2022', figsize=(15, 6))
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(20, 10))
axes = axes.flatten()

for i, col in enumerate(columns1):
    ax = axes[i]
    new_df[col].plot(kind='density', ax=ax, title=col)
    ax.axvline(new_df[col].mean(), color='red', linestyle='--', linewidth=1)
    ax.axvline(new_df[col].median(), color='green', linestyle='--', linewidth=1)
    ax.set_xlabel("Temp Change (Â°C)")
    ax.set_ylabel("Density")

# Remove any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(20, 10))
axes = axes.flatten()

for i, col in enumerate(columns2):
    ax = axes[i]
    new_df[col].plot(kind='density', ax=ax, title=col)
    ax.axvline(new_df[col].mean(), color='red', linestyle='--', linewidth=1)
    ax.axvline(new_df[col].median(), color='green', linestyle='--', linewidth=1)
    ax.set_xlabel("Temp Change (Â°C)")
    ax.set_ylabel("Density")

# Remove any unused subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Data Cleaning**

```python
# Impute NAs with the median of each column
df_imputed = new_df.fillna(round(new_df.median(numeric_only=True), 3))
df_imputed.isna().sum()
```

```python
# New dataframe with imputed values
df_imputed
```

```python
# Check for any remaining NA values
na_rows = df_imputed[df_imputed.isna().any(axis=1)]
na_rows
```

```python
# World is not a country, so it can be removed from the DataFrame
df_imputed = df_imputed[df_imputed['Country'] != 'World']
df_imputed
```

**Web Scraping**

```python
# Import libraries for web scraping
from bs4 import BeautifulSoup
import requests
```

```python
# Target URL
url = "https://www.countrycode.org/"

page = requests.get(url)

soup = BeautifulSoup(page.text, "html")
```

```python
# Print the soup object to see the HTML content
print(soup)
```

```python
# Find the table title
data_title = soup.find_all("th")[0:6]
data_title
```

```python
# Table title extraction
table_title = [title.text.strip() for title in data_title]
print(table_title)
```

```python
# Data extraction from the table
column_data = soup.find_all("td")
column_data
```

```python
# Data extracted into a list
column_data = [data.text.strip() for data in column_data]
column_data
```

```python
# Make sure the data is in chunk format
chunk_size = 6

rows = [column_data[i:i+chunk_size] for i in range(0, len(column_data), chunk_size)]
```

```python
# Store the data in a DataFrame
df_country = pd.DataFrame(rows, columns=table_title)
df_country
```

```python
# Drop unnecessary columns
df_country = df_country.drop(columns=['COUNTRY CODE', 'POPULATION', 'AREA KM2', 'GDP $USD'])
```

```python
# Display the remaining columns
df_country
```

```python
# Split the 'ISO CODES' column into 'ISO1' and 'ISO2'
df_country[['ISO1', 'ISO2']] = df_country['ISO CODES'].str.split('/', expand=True)
```

```python
# Final  DataFrame with relevant columns
df_country
```

**Merge Dataset**

```python
# Read the country codes CSV file
df_countryCode = pd.read_csv("country_codes.csv")
```

```python
# Remove leading and trailing spaces from ISO codes
df_imputed['ISO3'] = df_imputed['ISO3'].str.strip()
df_countryCode['ISO2'] = df_countryCode['ISO2'].str.strip()
```

```python
# Merge 2 DataFrames on ISO codes
complete_df = pd.merge(df_imputed, df_countryCode, left_on='ISO3', right_on='ISO2', how='left')
complete_df
```

```python
# Drop duplicate
complete_df = complete_df.drop_duplicates()
complete_df
```

```python
# Check for any remaining NA values in the merged DataFrame
na_rows = complete_df[complete_df.isna().any(axis=1)]
na_rows
```

```python
# Fill missing ISO2 values with ISO1 values
complete_df['ISO2_x'] = complete_df['ISO2_x'].fillna(complete_df['ISO1'])
```

```python
# Standardize the Country column
complete_df['Country'] = complete_df['COUNTRY'].combine_first(complete_df['Country'])
```

```python
# Drop unnecessary columns
complete_df = complete_df.drop(columns=['COUNTRY', 'ISO CODES', 'ISO1', 'ISO2_y'])
```

```python
# Rename columns
complete_df = complete_df.rename(columns={
    'ISO2_x': 'ISO1',
    'ISO3': 'ISO2'
})
complete_df
```
