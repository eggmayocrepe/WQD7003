```python
# Summarising Columns in Dataset
def Summary(df):
    total = len(df)

    print("\n=== Dataset Column Summary ===\n")
    for col in df.columns:
        missing_count = df[col].isnull().sum()
        missing_pct = (missing_count / total) * 100
        unique_vals = df[col].nunique(dropna=True)
        sample_vals = df[col].dropna().unique()[:5]  # preview up to 5 unique values

        print(f"--- {col} ---")
        print(f"Type: {df[col].dtype}")
        print(f"Missing: {missing_count} ({missing_pct:.2f}%)")
        print(f"Unique Values: {unique_vals}")
        print(f"Sample Unique Values: {sample_vals}")
        print("-" * 40)

# Summary function for a column
def summarise_column(df, col):
    if col in df.columns:
        total = len(df)
        non_churn_count = df[df[col] == 'Non_Churn'].shape[0]
        total_churn = total - non_churn_count
        non_churn_pct = (non_churn_count / total) * 100
        unique_vals = df[col].nunique()
        top_vals = df[col].value_counts()

        print(f"\n--- Summary of '{col}' Column ---")
        print(f"Non_Churn: {non_churn_count} out of {total} rows ({non_churn_pct:.2f}%)")
        print(f"Unique Values: {unique_vals}")

        print("\nValue Counts and Percentages (excluding 'Non_Churn'):")
        for value, count in top_vals.items():
            if value != 'Non_Churn':
                value_pct = (count / total_churn) * 100
                print(f"{value}: {count} ({value_pct:.2f}%)")
    else:
        print(f"Column '{col}' not found in DataFrame.")

Summary(df)
summarise_column(df, 'Churn Reason')
```

# Step 2: Handling Churn Column

```python
def group_reasons(reason):
    if reason == "Non_Churn":
        return "Non_Churn"
    reason = reason.lower()
    if any(keyword in reason for keyword in ['competitor']):
        return 'Competitor Offer'
    elif any(keyword in reason for keyword in ['attitude', 'expertise']):
        return 'Customer Service'
    elif any(keyword in reason for keyword in ['price', 'charge', 'cost', 'affordable', 'charges']):
        return 'Cost-related'
    elif 'network' in reason:
        return 'Network Issue'
    elif "don't know" in reason:
        return 'Uncertain'
    elif any(keyword in reason for keyword in ['service dissatisfaction', 'product dissatisfaction', 'limited', 'lack']):
        return 'Product Dissatisfaction'
    elif any(keyword in reason for keyword in ['moved', 'deceased']):
        return 'Personal Reasons'
    else:
        return 'Other'

# Apply grouping
df['Churn Reason Grouped'] = df['Churn Reason'].apply(group_reasons)

# Summarise the new grouped column
summarise_column(df, 'Churn Reason Grouped')

# Filter out 'Non_Churn'
filtered_df = df[df["Churn Reason Grouped"] != "Non_Churn"]

# Get group order (reversed)
group_order = filtered_df["Churn Reason Grouped"].value_counts().iloc[::-1].index

# Palette for distinct churn reasons
palette = sns.color_palette("tab20", n_colors=filtered_df["Churn Reason"].nunique())

# Plot
plt.figure(figsize=(14, 8))
ax = sns.countplot(
    data=filtered_df,
    y="Churn Reason Grouped",
    hue="Churn Reason",
    order=group_order,
    palette=palette,
    dodge=False,
)

# Title and axis labels
plt.title("Churn Reason vs Churn Reason Grouped", fontsize=16)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Churn Reason Grouped", fontsize=12)

# Legend formatting
handles, labels = ax.get_legend_handles_labels()
seen = set()
custom_handles = []
custom_labels = []

for group in group_order:
    reasons = filtered_df[filtered_df["Churn Reason Grouped"] == group]["Churn Reason"].unique()
    for reason in reasons:
        if reason not in seen and reason in labels:
            seen.add(reason)
            idx = labels.index(reason)
            custom_handles.append(handles[idx])
            custom_labels.append(f"{reason} ({group})")

plt.legend(
    custom_handles,
    custom_labels,
    title="Churn Reason (Group)",
    bbox_to_anchor=(1.05, 1),
    loc="upper left",
    fontsize=10,
)
plt.tight_layout()
plt.show()

# This could be handled by Churn Category, previously we didn't have this column. We've kept this for feature analysis.
```

3B. Categoric Columns

```python
def plot_stacked_bar_crosstab(df, index_col, column_col, title, xlabel='Category', ylabel='Percentage'):
    if index_col in df.columns and column_col in df.columns:
        # Create the crosstab
        cross_tab = pd.crosstab(df[index_col], df[column_col])

        # Normalise to get percentages
        cross_tab_pct = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100

        # Plot stacked bar chart with percentages
        ax = cross_tab_pct.plot(kind='bar', stacked=True, figsize=(9, 6), colormap='tab20c')

        # Add actual counts as text inside the bars
        for i, idx in enumerate(cross_tab.index):
            cumulative = 0
            for col in cross_tab.columns:
                height = cross_tab_pct.loc[idx, col]
                count = cross_tab.loc[idx, col]
                if height > 0:
                    ax.text(i, cumulative + height / 2, f'{count}\n({height:.1f}%)',
                            ha='center', va='center', fontsize=7, color='black')
                    cumulative += height

        plt.title(title, fontsize=12)
        plt.xlabel(xlabel, fontsize=10)
        plt.ylabel(ylabel, fontsize=10)
        plt.xticks(rotation=45, ha='right', fontsize=9)
        plt.yticks(fontsize=9)
        plt.legend(title=column_col, bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Columns '{index_col}' or '{column_col}' not found in DataFrame.")

# Apply the plot function to each column
for col in categoric_cols:
    if col != 'Churn Reason Grouped':
        plot_stacked_bar_crosstab(
            df,
            index_col='Churn Reason Grouped',
            column_col=col,
            title=f'{col} Distribution by Predicted Churn Reason',
            xlabel='Predicted Churn Reason'
        )
```

# **Step 6: Predictive Modelling**

Attempting to predict Churn.

```python
# --- Selecting the best variables for Churn Predicition & Analysis ---
selected_cols = [
    'Tenure in Months',          # Tenure duration
    'Churn Score',               # Predictive churn score
    'CLTV',                      # Lifetime value
    'Satisfaction Score',        # Customer satisfaction
    'Total Extra Data Charges',  # Extra Charges
    'Total Charges',             # Total Charges
    'Age',                       # Age
    'Contract',                  # Contract type (categorical)
    'Payment Method',            # Payment type (categorical)
    'Gender',                    # Gender (categorical)
    'Married',                   # Marital status (categorical)
    'Dependents',                # Dependents (categorical)
    'Internet Type'              # Internet type (categorical)
]

# Filter df to include only the selected columns
df_selected = df[selected_cols]

# Automatically select numerical and categorical columns
numerical_cols = df_selected.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df_selected.select_dtypes(include=['object']).columns.tolist()
```

```python
# Select features and target
X = df[['Total Extra Data Charges', 'Total Charges', 'Payment Method', 'Satisfaction Score', 'Contract', 'Churn Score', 'Age']]
y = df['Churn Value']

# Convert numeric columns automatically
X = X.apply(pd.to_numeric, errors='coerce')

# Define categorical and numerical columns dynamically
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['number']).columns.tolist()

# Define preprocessing pipelines
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean'))
])
preprocessor = ColumnTransformer([
    ('cat', categorical_pipeline, categorical_cols),
    ('num', numerical_pipeline, numerical_cols)
])

# Models to test
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=34),
    'Random Forest': RandomForestClassifier(random_state=34, class_weight='balanced'),
    'Decision Tree': DecisionTreeClassifier(random_state=34, class_weight='balanced'),
    'SVM': SVC(probability=True, class_weight='balanced', random_state=34),
    'KNN': KNeighborsClassifier(),
    'XGBoost': XGBClassifier(eval_metric='mlogloss', use_label_encoder=False, random_state=34)
}

# Evaluation setup
results = {}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=34)
f1_macro = make_scorer(f1_score, average='macro')

# Evaluate each model using SMOTE
for name, model in models.items():
    clf = ImbPipeline(steps=[
        ('preprocess', preprocessor),
        ('smote', SMOTE(random_state=34)),
        ('classifier', model)
    ])
    f1_scores = cross_val_score(clf, X, y, cv=cv, scoring=f1_macro)
    acc_scores = cross_val_score(clf, X, y, cv=cv, scoring='accuracy')

    results[name] = {
        'f1_mean': f1_scores.mean(),
        'f1_std': f1_scores.std(),
        'acc_mean': acc_scores.mean(),
        'acc_std': acc_scores.std()
    }
    print(f"{name}:\n  F1 (macro) = {f1_scores.mean():.2f} (+/- {f1_scores.std():.2f})\n  Accuracy = {acc_scores.mean():.2f} (+/- {acc_scores.std():.2f})\n")

# Barplot for comparison
labels = list(results.keys())
f1s = [results[m]['f1_mean'] for m in labels]
f1_err = [results[m]['f1_std'] for m in labels]

# Ensure the lengths match for f1s and f1_err
assert len(f1s) == len(f1_err), "f1s and f1_err must have the same length"

# Create a bar plot with error bars using plt.bar
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, f1s, yerr=f1_err, capsize=5, color='lightblue', edgecolor='black')
plt.title('Model Comparison: Macro F1 Score')
plt.ylabel('Mean F1 Score')
plt.xticks(rotation=45)
plt.grid(True)

# Add error bar caps
for bar in bars:
    bar.set_edgecolor('black')

plt.tight_layout()
plt.show()
```
